```python
#275
news='https://news.daum.net/'
soup=bs(ur.urlopen(news).read(),'html.parser')
headline=soup.find_all('div',{'class':'item_issue'})
for i in headline:
    print(i.text,'\n') #headline에서  <div> 태그를 하나씩 가져옵니다.
    b=i.find_all('a')[0].get('href')#headline기사의 url
    print(b) 
    soup3=bs(ur.urlopen(b).read(),'html.parser')  #i.find_all('a')[0].get('href') 해당기사의 url
    for j in soup3.find_all('p'):
        print(j.text)       
```
