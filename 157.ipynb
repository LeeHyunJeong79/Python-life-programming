```python
#160
import os,re
import usecsv
apt=usecsv.switch(usecsv.opencsv('apt_202109.csv'))
new_list=[]
for i in apt:
    try:
        if i[5]>=84 and i[-5]<=200000 and i[-5]>=150000 and re.match('경기',i[0]):
            #면적 84이상, 가격 20억~12억, 경기을 포함하는 조건
            new_list.append([i[0],i[4],i[-5]])
            #시군구, 아파트 단지명, 가격을 출력합니다.
    except :
        pass
print(new_list)
usecsv.writecsv('apt202109_2.csv',new_list)
```
