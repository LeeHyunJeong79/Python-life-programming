```python
#129p
import csv
def opencsv(filename): #csv파일 읽기
    f=open(filename,'r')
    reader=csv.reader(f)
    output=[]
    for i in reader:
        output.append(i)
    return output
    f.close()
opencsv('seoul.csv')
```

```python
def writecsv(filename,the_list):
    f=open(filename,'w',newline='')
    a=csv.writer(f,delimiter=',')
    a.writerows(the_list)
    f.close()
a=[['구','전체','내국인','외국인'],['관악구','5198','5030','1777'],['강남구','5476','5424','5104']]
writecsv('a.csv',a)
```
